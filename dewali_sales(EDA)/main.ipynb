```python
df=pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Cust_name','Status','unnamed1'],inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset='Amount',inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.set_style('whitegrid')
sns.boxplot(x='Orders',y='Amount',data=sales)
```

```python
# Outlier analysis 
Orders = df['Orders']

# # Calculate Q1 and Q3
Q1 = Orders.quantile(0.25)
Q3 = Orders.quantile(0.75)
IQR = Q3 - Q1

# # Filter out outliers
lowerfence = Q1 - (1.5*IQR)
higherfence = Q3 + (1.5*IQR)
# Filter out outliers

df_no_outliers = df[(Orders >= lowerfence) & (Orders <= higherfence)]
df = df_no_outliers
df.head()
```

```python
sns.set_style('whitegrid')
sns.boxplot(x='Orders',y='Amount',data=df_no_outliers)
```

```python
Y = pd.DataFrame(df['Amount']) # dependent features 
Y.head()
```

```python
X = pd.DataFrame(df.drop(columns=['Amount','Age Group']))
X.head()
```

```python
X.isnull().sum()
X.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score
from sklearn.metrics import mean_squared_error,r2_score
linreg = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1)
linreg.fit(X_train,y_train)
Y_pred = linreg.predict(X_test)
mean_ypred = Y_pred.mean()
print(mean_ypred)
mean_actual = df.Amount.mean()
print(mean_actual)
```

```python
mean_squared_error(y_test,Y_pred)
```

```python
r2_score(y_test,Y_pred)
```

```python
cross_score = cross_val_score(linreg, X_train, y_train , scoring='neg_mean_squared_error', cv=5) #for cross validation mush 
mean_cross_val = cross_score.mean()
mean_cross_val
```
